Importing the necessary modules needed for making the NN model

```python
from sklearn.model_selection import train_test_split
from keras.layers import Dense, Activation
from keras.models import Sequential
from keras.optimizers import SGD, Adam, RMSprop
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn import preprocessing
import pandas as pd
import numpy as np

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

The final data file is comprised from data from two sources
- The CDC database by runing a request for CSV based on state, year and cause of death firearm
- The State Firearm Laws Project codebook was used to get data on the state laws pertaining to Gun Restrictions
Function to read in CSV file

```python
def readData():
    data = pd.read_csv('raw_data (1).csv')
    data = data.drop(['state','year'], axis=1)
    return data
```

Function to normalize all the columns of dataframe

```python
def normalizeData(df):
    x = df.values
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled = min_max_scaler.fit_transform(x)
    data = pd.DataFrame(x_scaled)
    data.columns = df.columns
    return data
```

Function to predict accuracy score. If predicted value exactly matches actual value score increases by 1. Default score = 0

```python
def accuracy_score(y_preds, y_true):
    score = 0
    for i in range(len(y_preds)):
        print(y_preds[i], "  ",  y_true[i])
        if(y_preds[i] == y_true[i]):
            score += 1
    return score/len(y_preds)
```

Defining the nueral net
- Sequential model is used for defining nueral net
- One input layer with 49 nodes, 3 hidden layers with 25, 12 and 6 defining the input shape respectively for each layer
  and finally 1 output layer with one node with linear combination of inputs to form output
- Activation function used is relu - rectified linear unit for changing the activation node values by gradient descent
- The model is paramterized to reduce mse
- The model is run for 1000 epochs

```python
def NeuralNetwork(X_train, Y_train, X_test, Y_test):
    model = Sequential()
    
    model.add(Dense(49, input_dim=X_train.shape[1]))
    model.add(Activation('relu'))
    model.add(Dense(25))
    model.add(Activation('relu'))
    model.add(Dense(12, input_dim=16))
    model.add(Activation('relu'))
    model.add(Dense(6, input_dim=8))
    model.add(Activation('relu'))
    model.add(Dense(1))
    model.add(Activation('linear'))
    
    sgd = SGD(lr=0.001, decay=1e-2, momentum=0.3, nesterov=True)
    adam = Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
    rmsprop = RMSprop(lr=0.001, rho=0.5, epsilon=None, decay=0.0)
    model.compile(loss='mse', optimizer=sgd, metrics=['accuracy'])
    
    model.fit(X_train, Y_train, epochs=1000, batch_size=1, verbose=1)
    score = model.evaluate(X_test, Y_test, batch_size=1)
    preds = model.predict(X_test, batch_size=1, verbose=0)
    print(score)
    res = accuracy_score(preds, Y_test)
    print('SCORE : ', res)
```

Train  = 80% , Test 20% 

```python
def main():
    data = readData()
    data = normalizeData(data)
    # print(data.columns.values)
    train, test = train_test_split(data, test_size=0.2)
    X_train = train.values[:,1:]
    X_test = test.values[:,1:]
    Y_train = train.values[:,:1]
    Y_test = test.values[:,:1]
    NeuralNetwork(X_train, Y_train, X_test, Y_test)
```

```python
if __name__ == '__main__':
    main()
```
